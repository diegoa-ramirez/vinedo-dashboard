```python
# ==============================================
# üåê Servidor Web de Monitoreo de Vi√±edo
# Visualizaci√≥n en tiempo real de par√°metros simulados
# ==============================================

!pip install flask==2.2.5 pyngrok --quiet

from flask import Flask, jsonify, render_template_string
from pyngrok import ngrok
import random, threading, time

# --- Inicializaci√≥n del servidor Flask ---
app = Flask(__name__)

# --- Variables globales para los datos ---
datos_actuales = {
    "humedad_suelo": 0,
    "temperatura_suelo": 0,
    "ph_suelo": 0,
    "humedad_aire": 0,
    "temperatura_aire": 0,
    "radiacion": 0
}

# --- Funci√≥n para simular los datos continuamente ---
def actualizar_datos():
    while True:
        datos_actuales["humedad_suelo"] = round(random.uniform(10, 40), 2)
        datos_actuales["temperatura_suelo"] = round(random.uniform(15, 35), 2)
        datos_actuales["ph_suelo"] = round(random.uniform(5.5, 8.0), 2)
        datos_actuales["humedad_aire"] = round(random.uniform(40, 90), 2)
        datos_actuales["temperatura_aire"] = round(random.uniform(15, 35), 2)
        datos_actuales["radiacion"] = round(random.uniform(0, 50000), 2)
        time.sleep(2)

# --- Hilo en paralelo para la simulaci√≥n de datos ---
threading.Thread(target=actualizar_datos, daemon=True).start()

# --- Plantilla HTML (interfaz web) ---
html = """
<html>
<head>
  <title>üåø Monitoreo del Vi√±edo - Robot M√≥vil</title>
  <meta http-equiv="refresh" content="2">
  <style>
    body { font-family: Arial; background-color: #f0f7f0; color: #222; margin: 40px; }
    h1 { color: #2a6b2d; }
    ul { font-size: 18px; }
    .container { background: #fff; border-radius: 10px; padding: 20px; width: 400px;
                 box-shadow: 0 0 10px rgba(0,0,0,0.2); }
  </style>
</head>
<body>
  <h1>üçá Monitoreo del Vi√±edo</h1>
  <h3>Datos simulados en tiempo real</h3>
  <div class="container">
    <ul>
      <li><b>Humedad del suelo:</b> {{h['humedad_suelo']}} %</li>
      <li><b>Temperatura del suelo:</b> {{h['temperatura_suelo']}} ¬∞C</li>
      <li><b>pH del suelo:</b> {{h['ph_suelo']}}</li>
      <li><b>Humedad del aire:</b> {{h['humedad_aire']}} %</li>
      <li><b>Temperatura del aire:</b> {{h['temperatura_aire']}} ¬∞C</li>
      <li><b>Radiaci√≥n solar:</b> {{h['radiacion']}} lux</li>
    </ul>
  </div>
  <p><i>Actualizado autom√°ticamente cada 2 segundos</i></p>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(html, h=datos_actuales)

# --- Exponer el servidor con ngrok ---
print("‚è≥ Iniciando servidor Flask y creando t√∫nel p√∫blico...")
public_url = ngrok.connect(5000)
print("\n‚úÖ Servidor web en l√≠nea.")
print("üì° Puedes ver los datos del vi√±edo en tiempo real en:")
print(f"üëâ {public_url}")
print("\n(Abre este enlace en tu navegador o comp√°rtelo para acceder desde cualquier dispositivo)\n")

app.run(port=5000)
```

```python
# C√≥digo integrado: Flask-SocketIO + ngrok + simulador que emite datos en tiempo real
!pip install flask-socketio eventlet flask pyngrok chart-studio --quiet

from flask import Flask, render_template_string
from flask_socketio import SocketIO
from pyngrok import ngrok
import random, threading, time
import eventlet
eventlet.monkey_patch()

# ---------------------------
# CONFIG
# ---------------------------
PORT = 5000
EMIT_INTERVAL = 1.0  # segundos entre lecturas
# Umbrales (ejemplo)
UMBRAL_HUMEDAD_SUELO = 35
UMBRAL_TEMP_SUELO = 33
UMBRAL_PH_SUELO = 7.5
UMBRAL_HUMEDAD_AIRE = 85
UMBRAL_TEMP_AIRE = 35
UMBRAL_RADIACION = 45000

# ---------------------------
# App y SocketIO
# ---------------------------
app = Flask(__name__)
app.config['SECRET_KEY'] = 'secret!'
socketio = SocketIO(app, cors_allowed_origins="*", async_mode='eventlet')

# ---------------------------
# Datos compartidos / almacenamiento hist√≥rico (opcional)
# ---------------------------
hist = {
    "humedad_suelo": [],
    "temperatura_suelo": [],
    "ph_suelo": [],
    "humedad_aire": [],
    "temperatura_aire": [],
    "radiacion": [],
    "alertas": []
}

# ---------------------------
# Simulador (hilo)
# ---------------------------
def generar_datos():
    return {
        "humedad_suelo": round(random.uniform(10, 40), 2),
        "temperatura_suelo": round(random.uniform(15, 35), 2),
        "ph_suelo": round(random.uniform(5.5, 8.0), 2),
        "humedad_aire": round(random.uniform(40, 90), 2),
        "temperatura_aire": round(random.uniform(15, 35), 2),
        "radiacion": round(random.uniform(0, 50000), 2)
    }

def verificar_alertas(datos):
    alertas = []
    if datos["humedad_suelo"] > UMBRAL_HUMEDAD_SUELO: alertas.append("Humedad Suelo")
    if datos["temperatura_suelo"] > UMBRAL_TEMP_SUELO: alertas.append("Temp Suelo")
    if datos["ph_suelo"] > UMBRAL_PH_SUELO: alertas.append("pH Suelo")
    if datos["humedad_aire"] > UMBRAL_HUMEDAD_AIRE: alertas.append("Humedad Aire")
    if datos["temperatura_aire"] > UMBRAL_TEMP_AIRE: alertas.append("Temp Aire")
    if datos["radiacion"] > UMBRAL_RADIACION: alertas.append("Radiaci√≥n")
    return alertas

def simulador_emit():
    print("Simulador iniciado: generando datos y emitiendo por Socket.IO cada", EMIT_INTERVAL, "seg.")
    try:
        while True:
            datos = generar_datos()
            alertas = verificar_alertas(datos)
            datos['alertas'] = alertas  # a√±adimos lista de alertas al payload

            # Guardar en hist√≥rico (opcional)
            hist["humedad_suelo"].append(datos["humedad_suelo"])
            hist["temperatura_suelo"].append(datos["temperatura_suelo"])
            hist["ph_suelo"].append(datos["ph_suelo"])
            hist["humedad_aire"].append(datos["humedad_aire"])
            hist["temperatura_aire"].append(datos["temperatura_aire"])
            hist["radiacion"].append(datos["radiacion"])
            hist["alertas"].append(", ".join(alertas) if alertas else "")

            # Emitir a todos los clientes conectados
            socketio.emit('sensor_update', datos)

            # espera
            socketio.sleep(EMIT_INTERVAL)
    except Exception as e:
        print("Simulador finalizado:", e)

# Lanza el simulador en hilo daemon
threading.Thread(target=simulador_emit, daemon=True).start()

# ---------------------------
# P√°gina web (HTML + JS con Socket.IO client + Chart.js)
# ---------------------------
HTML = """
<!doctype html>
<html>
  <head>
    <meta charset="utf-8" />
    <title>Monitoreo Vi√±edo (Socket.IO)</title>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/socket.io/4.5.4/socket.io.min.js"></script>
    <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
    <style>
      body { font-family: Arial, sans-serif; margin:20px; background:#f6fbf6; color:#222;}
      h1 { color:#2a6b2d; }
      .grid { display:grid; grid-template-columns: repeat(2, 1fr); gap:20px; max-width:1100px; }
      .card { background:white; padding:12px; border-radius:8px; box-shadow:0 2px 6px rgba(0,0,0,0.08);}
      .small { font-size:14px; color:#555; margin-top:6px; }
    </style>
  </head>
  <body>
    <h1>üçá Monitoreo en tiempo real - Robot m√≥vil</h1>
    <p>Conexi√≥n por WebSocket. Actualiza autom√°ticamente.</p>
    <div class="grid">
      <div class="card"><canvas id="chart_humedad_suelo"></canvas><div class="small">Humedad Suelo (%)</div></div>
      <div class="card"><canvas id="chart_temp_suelo"></canvas><div class="small">Temperatura Suelo (¬∞C)</div></div>
      <div class="card"><canvas id="chart_ph_suelo"></canvas><div class="small">pH Suelo</div></div>
      <div class="card"><canvas id="chart_humedad_aire"></canvas><div class="small">Humedad Aire (%)</div></div>
      <div class="card"><canvas id="chart_temp_aire"></canvas><div class="small">Temperatura Aire (¬∞C)</div></div>
      <div class="card"><canvas id="chart_radiacion"></canvas><div class="small">Radiaci√≥n (lux)</div></div>
    </div>

    <h3>√öltima lectura:</h3>
    <pre id="last_reading" style="background:#fff;padding:10px;border-radius:6px;max-width:1100px;"></pre>

    <script>
      const socket = io();

      // Crear arrays iniciales
      const maxPoints = 60;
      const dataStore = {
        hs: [], ts: [], ph: [], ha: [], ta: [], r: [], xlabels: []
      };

      // Crear charts
      function createChart(ctx, label, color) {
        return new Chart(ctx, {
          type: 'line',
          data: {
            labels: dataStore.xlabels,
            datasets: [{
              label: label,
              data: [],
              fill: false,
              borderWidth: 2,
              tension: 0.2
            }]
          },
          options: {
            animation: false,
            responsive: true,
            plugins: { legend: { display: false } },
            scales: {
              x: { display: true, title: { display:false } },
              y: { display: true, title: { display: true, text: label } }
            }
          }
        });
      }

      const ch_hs = createChart(document.getElementById('chart_humedad_suelo').getContext('2d'), 'Humedad Suelo (%)');
      const ch_ts = createChart(document.getElementById('chart_temp_suelo').getContext('2d'), 'Temp Suelo (¬∞C)');
      const ch_ph = createChart(document.getElementById('chart_ph_suelo').getContext('2d'), 'pH Suelo');
      const ch_ha = createChart(document.getElementById('chart_humedad_aire').getContext('2d'), 'Humedad Aire (%)');
      const ch_ta = createChart(document.getElementById('chart_temp_aire').getContext('2d'), 'Temp Aire (¬∞C)');
      const ch_r  = createChart(document.getElementById('chart_radiacion').getContext('2d'), 'Radiaci√≥n (lux)');

      function pushData(value, array) {
        array.push(value);
        if (array.length > maxPoints) array.shift();
      }

      socket.on('sensor_update', function(datos) {
        const t = new Date().toLocaleTimeString();
        dataStore.xlabels.push(t);
        if (dataStore.xlabels.length > maxPoints) dataStore.xlabels.shift();

        pushData(datos.humedad_suelo, dataStore.hs);
        pushData(datos.temperatura_suelo, dataStore.ts);
        pushData(datos.ph_suelo, dataStore.ph);
        pushData(datos.humedad_aire, dataStore.ha);
        pushData(datos.temperatura_aire, dataStore.ta);
        pushData(datos.radiacion, dataStore.r);

        // actualizar datasets
        ch_hs.data.labels = dataStore.xlabels; ch_hs.data.datasets[0].data = dataStore.hs; ch_hs.update();
        ch_ts.data.labels = dataStore.xlabels; ch_ts.data.datasets[0].data = dataStore.ts; ch_ts.update();
        ch_ph.data.labels = dataStore.xlabels; ch_ph.data.datasets[0].data = dataStore.ph; ch_ph.update();
        ch_ha.data.labels = dataStore.xlabels; ch_ha.data.datasets[0].data = dataStore.ha; ch_ha.update();
        ch_ta.data.labels = dataStore.xlabels; ch_ta.data.datasets[0].data = dataStore.ta; ch_ta.update();
        ch_r .data.labels = dataStore.xlabels; ch_r .data.datasets[0].data = dataStore.r;  ch_r.update();

        // Mostrar √∫ltimo JSON
        document.getElementById('last_reading').textContent = JSON.stringify(datos, null, 2);
      });
    </script>
  </body>
</html>
"""

@app.route('/')
def index():
    return render_template_string(HTML)

# ---------------------------
# Exponer con ngrok y ejecutar
# ---------------------------
print("Creando t√∫nel p√∫blico con ngrok...")
public_url = ngrok.connect(PORT)
print("\n‚úÖ Servidor en l√≠nea.")
print("üì° URL p√∫blica:", public_url)
print("Abre esa URL en un navegador para ver el dashboard en tiempo real.\n")

# Ejecutar SocketIO (bloqueante)
try:
    socketio.run(app, host='0.0.0.0', port=PORT)
except KeyboardInterrupt:
    print("Servidor detenido por el usuario.")
finally:
    print("Finalizando.")
    # Aqu√≠ podr√≠as guardar hist en Excel si quieres exportar al detener
    # por ejemplo: crear un DataFrame con pandas y usar to_excel
```

```python
# ==============================================================
#  SIMULACI√ìN DE TRANSMISI√ìN DE DATOS DEL VI√ëEDO HACIA THINGSPEAK
# ==============================================================

import random, time, requests
from datetime import datetime
import pandas as pd

# --------------------------------------------------------------
# CONFIGURACI√ìN DEL CANAL THINGSPEAK
# --------------------------------------------------------------
API_KEY = "TU_API_KEY_AQUI"  # <-- Reemplaza esto con tu clave real
URL = f"https://api.thingspeak.com/update?api_key={API_KEY}"

# --------------------------------------------------------------
# LISTAS PARA GUARDAR DATOS LOCALMENTE (opcional)
# --------------------------------------------------------------
humedad_suelo, temperatura_suelo, ph_suelo = [], [], []
humedad_aire, temperatura_aire, radiacion = [], [], []

# --------------------------------------------------------------
# UMBRALES PARA ALERTAS
# --------------------------------------------------------------
UMBRAL_HUMEDAD_SUELO = 35
UMBRAL_TEMP_SUELO = 33
UMBRAL_PH_SUELO = 7.5
UMBRAL_HUMEDAD_AIRE = 85
UMBRAL_TEMP_AIRE = 35
UMBRAL_RADIACION = 45000

print("üåø Iniciando simulaci√≥n... (presiona Ctrl + C para detener)")
print("Los datos se enviar√°n a ThingSpeak cada 15 segundos.\n")

try:
    while True:
        # ----------------------------------------------------------
        # GENERACI√ìN DE DATOS SIMULADOS
        # ----------------------------------------------------------
        datos = {
            "humedad_suelo": round(random.uniform(10, 40), 2),
            "temperatura_suelo": round(random.uniform(15, 35), 2),
            "ph_suelo": round(random.uniform(5.5, 8.0), 2),
            "humedad_aire": round(random.uniform(40, 90), 2),
            "temperatura_aire": round(random.uniform(15, 35), 2),
            "radiacion": round(random.uniform(0, 50000), 2)
        }

        # ----------------------------------------------------------
        # ENV√çO DE DATOS A THINGSPEAK (HTTP GET)
        # ----------------------------------------------------------
        payload = {
            'field1': datos["humedad_suelo"],
            'field2': datos["temperatura_suelo"],
            'field3': datos["ph_suelo"],
            'field4': datos["humedad_aire"],
            'field5': datos["temperatura_aire"],
            'field6': datos["radiacion"]
        }

        response = requests.get(URL, params=payload)
        if response.status_code == 200:
            print(f"{datetime.now().strftime('%H:%M:%S')} ‚Üí Datos enviados correctamente.")
        else:
            print(f"‚ö†Ô∏è Error al enviar datos: {response.status_code}")

        # ----------------------------------------------------------
        # GUARDAR DATOS LOCALMENTE (opcional)
        # ----------------------------------------------------------
        humedad_suelo.append(datos["humedad_suelo"])
        temperatura_suelo.append(datos["temperatura_suelo"])
        ph_suelo.append(datos["ph_suelo"])
        humedad_aire.append(datos["humedad_aire"])
        temperatura_aire.append(datos["temperatura_aire"])
        radiacion.append(datos["radiacion"])

        df = pd.DataFrame({
            "Humedad Suelo": humedad_suelo,
            "Temperatura Suelo": temperatura_suelo,
            "pH Suelo": ph_suelo,
            "Humedad Aire": humedad_aire,
            "Temperatura Aire": temperatura_aire,
            "Radiaci√≥n": radiacion
        })
        df.to_csv("DatosLocales.csv", index=False)

        # ----------------------------------------------------------
        # DETECCI√ìN DE ALERTAS
        # ----------------------------------------------------------
        alertas = []
        if datos["humedad_suelo"] > UMBRAL_HUMEDAD_SUELO: alertas.append("Humedad Suelo")
        if datos["temperatura_suelo"] > UMBRAL_TEMP_SUELO: alertas.append("Temp Suelo")
        if datos["ph_suelo"] > UMBRAL_PH_SUELO: alertas.append("pH Suelo")
        if datos["humedad_aire"] > UMBRAL_HUMEDAD_AIRE: alertas.append("Humedad Aire")
        if datos["temperatura_aire"] > UMBRAL_TEMP_AIRE: alertas.append("Temp Aire")
        if datos["radiacion"] > UMBRAL_RADIACION: alertas.append("Radiaci√≥n")

        if alertas:
            print(f"‚ö†Ô∏è ALERTA: {', '.join(alertas)}\n")
        else:
            print("‚úÖ Todos los par√°metros dentro del rango normal.\n")

        # ----------------------------------------------------------
        # INTERVALO ENTRE ENV√çOS
        # ----------------------------------------------------------
        time.sleep(15)  # m√≠nimo 15 s entre env√≠os gratuitos

except KeyboardInterrupt:
    print("\nüõë Simulaci√≥n detenida por el usuario.")
    print("Los datos han sido almacenados localmente en 'DatosLocales.csv'")
```
